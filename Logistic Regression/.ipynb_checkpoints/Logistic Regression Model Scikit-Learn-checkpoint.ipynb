```python
# Universal
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Cleaning and Analysis
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
# Logistic Regression
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import auc, roc_curve
```

In this data each row represents a different employee, and the columns include information such as age, education, etc. In total, there are 14 features corresponding to the first 14 columns in the datasets. The last income column shows if each employee has an income â‰¤ 50K or > 50K, which represents true labels of the samples. The training set contains 26, 049 data samples and the test set contains 6, 512 samples.

```python
# Load Data
train_data = pd.read_csv('train_data.csv')
test_data = pd.read_csv('test_data.csv')
```

```python
# Clean Data
def clean_data(train_data, test_data):
    # fill in missing values
    train_data = train_data.replace(["?"], np.nan)
    test_data = test_data.replace(["?"], np.nan)
    frequent_imputer = SimpleImputer(missing_values=np.nan, strategy= 'most_frequent')
    train_data.iloc[:, 0:15] = frequent_imputer.fit_transform(train_data.iloc[:, 0:15])
    test_data.iloc[:, 0:15] = frequent_imputer.fit_transform(test_data.iloc[:, 0:15])
    
    # replace strange values
    train_data = train_data.replace([99999], np.nan)
    test_data = test_data.replace([99999], np.nan)
    mean_imputer = SimpleImputer(missing_values= np.nan , strategy='mean')
    train_data.iloc[:, 10:11] = mean_imputer.fit_transform(train_data.iloc[:, 10:11])
    test_data.iloc[:, 10:11] = mean_imputer.fit_transform(test_data.iloc[:, 10:11])
    
    # drop unnecessary columns aka education.num and fnlwgt
    test_data = test_data.drop(['fnlwgt', 'education.num'], axis=1, inplace=False)
    train_data = train_data.drop(['fnlwgt', 'education.num'], axis=1, inplace=False)
    
    # simplify race and education columns
    train_data["race"] = train_data["race"].replace(["Black", "Asian-Pac-Islander", "Amer-Indian-Eskimo"], "Other")
    test_data["race"] = test_data["race"].replace(["Black", "Asian-Pac-Islander", "Amer-Indian-Eskimo"], "Other")
    train_data["education"] = train_data["education"].replace(["11th", "9th", "7th-8th", "5th-6th", "10th", "1st-4th", "Preschool", "12th"], "School")
    test_data["education"] = test_data["education"].replace(["11th", "9th", "7th-8th", "5th-6th", "10th", "1st-4th", "Preschool", "12th"], "School")
    
    # convert labels into binary values aka income column
    label_encoder = LabelEncoder()
    train_data.iloc[:, 12] = label_encoder.fit_transform(train_data.iloc[:, 12])
    test_data.iloc[:, 12] = label_encoder.fit_transform(test_data.iloc[:, 12])
    
    # save processed data into cleaned_train_data.csv and cleaned_test_data.csv
    train_data.to_csv("cleaned_train_data.csv", index=False)
    test_data.to_csv("cleaned_test_data.csv", index=False)
    
    return

clean_data(train_data, test_data)
clean_train = pd.read_csv('cleaned_train_data.csv')
clean_test = pd.read_csv('cleaned_test_data.csv')
clean_train.head(5)
```

```python
# Plot a heatmap showing the correlation among numeric features and true labels using the training set.

def plot_numeric_feature_correlation(train_data):
    data = train_data[['age', 'capital.gain', 'capital.loss', 'hours.per.week', 'income']].corr()
    hm = sns.heatmap(data = data, annot = True, cmap = "Greens")
    plt.title("Correlation between different features")
    plt.show()

plot_numeric_feature_correlation(clean_train)
```

The first thing I noticed was how the age and income categories have the strongest correlations with all other categories, the strong correlation between age and income makes sense given how incomes increase with experience. I did expect that capital gain and capital loss would be more negatively correlated because I would expect that individuals making capital gains will keep making money and more likely to keep making gains rather than losses, but I guess the overwhelming mode of 0 dilutes the possibility. In the same way that individuals making capital gains will keep making money I would also expect that individuals making a higher income have a higher opportunity for capital gain, which is shown in the highest correlated pair income and capital gain.

The more you work the more you make.

```python
# Preprocessing on the datasets 
def preprocess_data(X_train, X_test):
    # encode categorical features using OneHotEncoder class
    # standardize numerical features using StandardScaler()class
    # return processed training and test sets
    train_processed = pd.DataFrame(data = {"age": StandardScaler().fit_transform(X_train.iloc[:, 0:1]).T[0]})
    test_processed = pd.DataFrame(data = {"age": StandardScaler().fit_transform(X_test.iloc[:, 0:1]).T[0]})
    col = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']
    ohe = OneHotEncoder()
    for i in col:
        train_col = np.unique(X_train[i].values)
        test_col = np.unique(X_test[i].values)
        train = pd.DataFrame(data=ohe.fit_transform(X_train[i].values.reshape(-1, 1)).toarray(), columns=train_col)
        test = pd.DataFrame(data=ohe.fit_transform(X_test[i].values.reshape(-1, 1)).toarray(), columns=test_col)
        
        for f in train_col:
            title = i + "." + f
            train_processed[title] = train[f]
            
        for t in test_col:
            title = i + "." + t
            test_processed[title] = test[t]
            
    stnd_col = ["capital.gain", "capital.loss", "hours.per.week"]
    stnd_train = StandardScaler().fit_transform(X_train.iloc[:, 8:11]).T 
    stnd_test = StandardScaler().fit_transform(X_test.iloc[:, 8:11]).T
    for z in range(3):
        title = "standardized." + stnd_col[z]
        train_processed[title] = stnd_train[z]
        test_processed[title] = stnd_test[z]
    
    # might have to add columns of zeros for test data that doesnt contain all outcomes as train
    # columns would be {'native.country.Holand-Netherlands'(col 63),'workclass.Never-worked'(col 3),'workclass.Without-pay'(col 8)}
    test_processed.insert(3, "workclass.Never-worked", np.zeros((len(test_processed))))
    test_processed.insert(8, "workclass.Without-pay", np.zeros((len(test_processed))))
    test_processed.insert(63, "native.country.Holand-Netherlands", np.zeros((len(test_processed))))
    return train_processed, test_processed

X_train = clean_train.drop(['income'], axis=1, inplace=False)
y_train = clean_train['income']
X_test = clean_test.drop(['income'], axis=1, inplace=False)
y_test = clean_test['income']
processed_X_train, processed_X_test = preprocess_data(X_train, X_test)
processed_X_train.head(5)
```

```python
# Build and train a logistic regression model to predict if an employee has a salary higher than $50K 
# Implementing the gradient ascent method to train the mode

class LogisticRegression(object):
    def __init__(self, X_train, y_train, X_test, y_test, learn_rate=0.001, n_iters=500):
        if len(X_train.shape) == 1:
            self.X_train = np.hstack((np.array([1]), X_train))
        else:
            self.X_train = np.hstack((np.ones((X_train.shape[0], 1)), X_train))
        self.y_train = y_train

        if len(X_test.shape) == 1:
            self.X_test = np.hstack((np.array([1]), X_test))
        else:
            self.X_test = np.hstack((np.ones((X_test.shape[0], 1)), X_test))
        self.y_test = y_test

        self.learn_rate = learn_rate
        self.n_iters = n_iters 

        # Initialize model weights
        self.weight = np.zeros((self.X_train.shape[1]))

        self.accuracy_train = []  # To keep track of accuracy on the training data during gradient ascent process
        self.accuracy_test = []  # To keep track of accuracy on the test data during gradient ascent process

        self.log_likelihood_train = []  # To keep track of average log likelihood on the training data during gradient ascent process
        self.log_likelihood_test = []  # To keep track of average log likelihood on the test data during gradient ascent process

    def predict(self, samples, threshold=0.5):
        return self.predict_prob(samples) >= threshold

    def predict_prob(self, samples):
        # Compute and return the predicted probability of each sample having an income >= 50K
        z = np.dot(samples,self.weight)
        probs = (1/(1+np.exp(-z)))
        return probs

    def compute_gradient(self, y, X):
        # Compute and return gradient of the average log likelihood on the training set
        # y is a vector
        # X is feature matrix
        #predict_prob(X, w) is the probability of every row given the weights and a row of data
        y_hat = self.predict_prob(X)
        gradient = np.mean((y-y_hat)*X.T, axis=1)
        return gradient

    def gradient_ascent(self):
        # Run gradient ascent to learn model weights
        # Store the average log likelihood and the prediction accuracy 
        # on the training and test sets after every gradient ascent iteration
        # need y vector and X array(featuresxexamples)
        
        for i in range(self.n_iters):
            gradient = self.compute_gradient(self.y_train, self.X_train)
            self.weight +=  self.learn_rate*gradient
            
            #Calulating probs
            y_hat_train = self.predict_prob(self.X_train)
            y_hat_test = self.predict_prob(self.X_test)
            
            # Calculating log likelihood
            train_ll = self.log_likelihood(self.y_train, y_hat_train)
            self.log_likelihood_train.append(train_ll)
            
            test_ll = self.log_likelihood(self.y_test, y_hat_test)
            self.log_likelihood_test.append(test_ll)
            
            # Calculating Binary Values
            bin_y_hat_train = self.predict(self.X_train)
            bin_y_hat_test = self.predict(self.X_test)
            
            #Calculating Accuracy
            train_acc = self.accuracy(self.y_train, bin_y_hat_train)
            self.accuracy_train.append(train_acc)
            
            test_acc = self.accuracy(self.y_test, bin_y_hat_test)
            self.accuracy_test.append(test_acc)
        return

            
    def log_likelihood(self, y_true, y_pred):
        # Fix 0/1 values in y_pred so that log is not undefined
        # Define epsilon because log(0) is not defined
        min_val = 1e-7
        y_pred = np.maximum(np.full(y_pred.shape, min_val), np.minimum(np.full(y_pred.shape, 1-min_val), y_pred))
        likelihood = np.mean(y_true*np.log(y_pred)+(1-y_true)*np.log(1-y_pred))
        return likelihood
    

    def accuracy(self, y, y_hat):
        #y_hat is binary values
        count=0
        diff = y_hat - y
        count = np.count_nonzero(diff==0)   
        acc = (count/len(y_hat))
        return acc 
    
```

```python
# Run gradient ascent with Logistic Regression for a learning rate of 0.75
learner = LogisticRegression(processed_X_train, y_train.values, processed_X_test, y_test.values,
                                 n_iters=250, learn_rate=0.75)
learner.gradient_ascent()
```

```python
# Run gradient ascent with Logistic Regression for a learning rate of 0.05
learner2 = LogisticRegression(processed_X_train, y_train.values, processed_X_test, y_test.values,
                                 n_iters=250, learn_rate=0.05)
learner2.gradient_ascent()
```

```python
# Plot the average log-likelihood curves for the training and test sets during the training process
def plot_log_likelihood_performance(log_likelihood_train, log_likelihood_test):
    # plot the average log likelihood curves of the training and test sets
    # during training process
    plt.plot(log_likelihood_train, c="r")
    plt.plot(log_likelihood_test, c="g")
    plt.legend(["Train Log Likelihood", "Test Log Likelihood"])
    plt.ylabel("Log Likelihood")
    plt.xlabel("Iterations")
    plt.title("Log Likelihood During Training")

plot_log_likelihood_performance(learner.log_likelihood_train, learner.log_likelihood_test)
```

```python
plot_log_likelihood_performance(learner2.log_likelihood_train, learner2.log_likelihood_test)
```

```python
# Plot the prediction accuracy for the training and test sets during the training process
def plot_accuracy_performance(accuracy_train, accuracy_test):
    # plot the prediction accuracy curves of the training and test sets
    # during the training process
    plt.plot(accuracy_train, c='r')
    plt.plot(accuracy_test, c='g')
    plt.legend(["Train Accuracy", "Test Accuracy"])
    plt.ylabel("Accuracy")
    plt.xlabel("Iterations")
    plt.title("Accuracy During Training")

plot_accuracy_performance(learner.accuracy_train, learner.accuracy_test)
```

```python
plot_accuracy_performance(learner2.accuracy_train, learner2.accuracy_test)
```

When the learning rate is 0.75 the accuracy converges much faster than when the learning rate is 0.05 although they both end up at about the same place after 250 iterations. The accuracy curve is much smoother when the learning rate is 0.75 as well which surprised me. When it comes to log likelihood it is much higher when the learning rate is 0.75, when the learning rate is at 0.05 the log likelihood for the test and train is closer which is great but the learning rate is too small to get as high as when the learning rate is 0.75

```python
#  plot the ROC curve showing the performance of the trained logistic regression model on the input training and test sets
def plot_roc(X_train, y_train, X_test , y_test, learner):
    # plot the ROC curves
    w = learner.weight

    bin_y_hat_train = learner.predict_prob(X_train)
    bin_y_hat_test = learner.predict_prob(X_test)
    
    train_fpr, train_tpr, _ = roc_curve(y_train, bin_y_hat_train)
    train_auc = auc(train_fpr, train_tpr)
    
    test_fpr, test_tpr, _ = roc_curve(y_test, bin_y_hat_test)
    test_auc = auc(test_fpr, test_tpr)
    
    print("Area Under the Curve for the Train Set:", train_auc)
    print("Area Under the Curve for the Test Set:", test_auc)
    
    plt.plot(train_fpr, train_tpr, color='r')
    plt.plot(test_fpr, test_tpr, color='g')
    plt.legend(["Train ROC", "Test ROC"])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title("ROC Curves")
    return

plot_roc(learner.X_train, learner.y_train, learner.X_test, learner.y_test, learner)
```

```python
plot_roc(learner2.X_train, learner2.y_train, learner2.X_test, learner2.y_test, learner2)
```

Once again when the learning rate is 0.05 the ROC curve does not converge as fast however the difference is not as extreme as the difference in accuracy plots. The area under the curve is still pretty high which is a good thing, but is still slightly smaller than when the learning
rate is 0.75. When the learning rate is 0.05 the curve is also much less smooth and there is a slightly larger difference between train aoc and test aoc.
