```python
import numpy as np
import torch
from torch import nn
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split

num_classes = 4
num_features = 2
random_seed = 1

# Create simple dataset of 10000 samples in which each data sample has only two features and belong to one of four classes. 
# X is multidimensional array with 10000 arrays of length 2 and y is single dimensional of length 10000
X, y = make_blobs(n_samples=10000, n_features=num_features, centers=num_classes, 
                  cluster_std=1.5, random_state=random_seed)

# Convert to tensor
X = torch.from_numpy(X).type(torch.float)
y = torch.from_numpy(y).type(torch.LongTensor)

# Split dataset into train/test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_seed)
```

```python
# Build and design simple model 
class SimpleModel(nn.Module):
    def __init__(self, input_size, output_size):
        # determine which layers will be used in the network model here
        mid_size = 10
        super().__init__()
        self.l1 = nn.Linear(input_size, mid_size)
        self.l2 = nn.Linear(mid_size, output_size)
        self.l3 = nn.Linear(5, output_size)
        self.tanh = nn.Tanh()
        self.sigmoid = nn.Sigmoid()
        self.relu = nn.ReLU()

    def forward(self, x):
        # input is data samples and the output is the model prediction
        return self.tanh(self.l2(self.sigmoid(self.l1(x))))
```

```python
# Run a complete training epoch of forward and backward pass to update the model
def train_step(model, X_train, y_train, loss_fn, optimizer, device):
    # initialize the model and determine which loss function and which optimizer to use
    # keep track of the loss and accuracy of the model on both the training and test sets after every training epoch
    # output of this method includes the train loss and train accuracy of the model after the model update
    model.train()
    y_logits = model(X_train)
    y_hat = torch.argmax(torch.softmax(y_logits, dim=1), dim=1)

    #loss
    loss = loss_fn(y_logits, y_train)
    # accuracy 
    acc = torch.eq(y_hat, y_train).sum().item() / len(y_hat)
    
    # Update step
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    return loss, acc
    
```

```python
# Evaluate the model on the test set
def evaluation_step(model, X_test, y_test, loss_fn, device):
    # output is the loss and accuracy of the model on the test set
    model.eval()
    y_logits = model(X_test)
    y_hat = torch.argmax(torch.softmax(y_logits, dim=1), dim=1)

    #loss
    loss = loss_fn(y_logits, y_test)
    # accuracy 
    acc = torch.eq(y_hat, y_test).sum().item() / len(y_hat)

    return loss, acc
    
```

```python
# Train model using the methods train step and evaluation step implemented
def train_simple_model(X_train, y_train, X_test, y_test, random_seed):
    # initialize the model and determine which loss function and which optimizer to use
    # tune the hyper-parameters including the model architecture, the learning rate, and the regularization parameter, etc 
    # based on the performance of the model on the validation set during the training process
    
    # Make device
    device = "cpu" 
    if torch.cuda.is_available(): 
        device = "cuda"
    else:
        if torch.backends.mps.is_available():
            device = "mps" 
    
    torch.manual_seed(random_seed) # do not change this

    # Train the model here

    X_train, y_train = X_train.to(device), y_train.to(device)
    X_test, y_test = X_test.to(device), y_test.to(device)
    
    input_size = X_train.shape[1]
    output_size = len(torch.unique(y_train))
    model = SimpleModel(input_size, output_size).to(device)
    loss_fn = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=0.2)

    train_l_holder = []
    train_a_holder = []
    test_l_holder = []
    test_a_holder = []

    for epoch in range(2000):
        train_loss, train_acc = train_step(model, X_train, y_train, loss_fn, optimizer, device)
        test_loss, test_acc = evaluation_step(model, X_test, y_test, loss_fn, device)
        train_l_holder.append(train_loss.item())
        train_a_holder.append(train_acc)
        test_l_holder.append(test_loss.item())
        test_a_holder.append(test_acc)

    return model, train_l_holder, train_a_holder, test_l_holder, test_a_holder
```

```python
# Intialize and train the model
model, train_losses, train_accuracies, test_losses, test_accuracies\
    = train_simple_model(X_train, y_train, X_test, y_test, random_seed)
```

```python
# Plot the accuracy curves of the model on the training and test sets during the training process
def plot_accuracy_performance(train_accuracies, test_accuracies):
    print("Train:", max(train_accuracies), "Test:", max(test_accuracies))
    plt.plot(train_accuracies, c='r')
    plt.plot(test_accuracies, c='g')
    plt.legend(["Train Accuracy", "Test Accuracy"])
    plt.ylabel("Accuracy")
    plt.xlabel("Epochs")
    plt.title("Accuracy During Training")

plot_accuracy_performance(train_accuracies, test_accuracies)
```

```python
# Plot the loss curves of the model on the training and test sets during the training process
def plot_loss_performance(train_losses, test_losses):
    print("Train:", max(train_losses), "Test:", max(test_losses))
    plt.plot(train_losses, c='r')
    plt.plot(test_losses, c='g')
    plt.legend(["Train Loss", "Test Loss"])
    plt.ylabel("Cross Entropy Loss")
    plt.xlabel("Epochs")
    plt.title("Loss During Training")

plot_loss_performance(train_losses, test_losses)
```

The model is 91% accurate with a loss of 1.4. The loss and accuracy curves converge very quickly, getting very close to the max values of 2000 epochs within 200 epochs. The accuracy and loss are very similar for both the train and test model so there isn't an overfitting issue.

```python
# Plot the decision boundary outcome of the trained model with the scatter distribution of the training dataset
def plot_decision_boundary(model, X_train, y_train):
    x_min, x_max = X_train[:, 0].min() - 0.1, X_train[:,0].max() + 0.1
    y_min, y_max = X_train[:, 1].min() - 0.1, X_train[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min,x_max, 100), np.linspace(y_min, y_max, 100))
    x_in = np.c_[xx.ravel(), yy.ravel()]
    x_in = torch.from_numpy(x_in).type(torch.float)
    
    model.eval()
    y_logits = model(x_in)
    y_hat = torch.argmax(torch.softmax(y_logits, dim=1), dim=1)

    plt.contourf(xx, yy, y_hat.reshape(xx.shape), cmap=plt.cm.RdYlBu, alpha=0.7 )
    plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color="red")
    plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color="orange")
    plt.scatter(X_train[y_train == 2][:, 0], X_train[y_train == 2][:, 1], color="lightblue")
    plt.scatter(X_train[y_train == 3][:, 0], X_train[y_train == 3][:, 1], color="blue")

plot_decision_boundary(model, X_train, y_train)
```
